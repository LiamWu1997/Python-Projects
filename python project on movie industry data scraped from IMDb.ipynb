```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use ('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # this will adjust the configuration of the plots



# Read in the data

df = pd.read_csv(r'C:\Users\User\OneDrive\Documents\data porfolio project\movies.csv')
```

```python
# look up the data
df.head()
```

```python
# check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format (col, pct_missing))
    
# confirmed no missing data
```

```python
# check for data types

df.dtypes
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# drop duplicates using drop_duplicates()

df['company'].drop_duplicates().sort_values(ascending=False)
df['company'].head()
```

```python
# hyphothesis:
# -budget should correlate with gross
# -company should correlate with gross
```

```python
# quick outlier check using boxplot
df.boxplot(column=['gross'])
```

```python
# create scatter plot of budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget')
plt.show()
```

```python
df.head()
```

```python
# use seaborn to create linear regression to determine the correlation | plotting budget vs gross

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"blue"}, line_kws={"color":"red"})
```

```python
# quantifying correlation
```

```python
#there are 3 statistical methods which are pearson, kendall, spearman
#pearson is the default method

df.corr() 
```

```python
# budget and gross correlation is high, hypothesis confirmed
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Quantitative Analysis of Movies Data')
plt.xlabel('Variables')
plt.ylabel('Variables')

plt.show()
```

```python
# now check for correlation between company and gross and other variables

df.head()
```

```python
# company is not numeric, so a numeric representation has to be created
# convert all fields into numeric value

df_categorized = df

for col_name in df_categorized.columns:
    if(df_categorized[col_name].dtype =='object'):
        df_categorized[col_name] = df_categorized[col_name].astype('category')
        df_categorized[col_name] = df_categorized[col_name].cat.codes
        
df_categorized.head()
```

```python
# generate a new correlation matrix with the categorized table

correlation_matrix = df_categorized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Quantitative Analysis of Movies Data')
plt.xlabel('Variables')
plt.ylabel('Variables')

plt.show()
```

```python
#now organize these so that the highest correlations are 
#use unstaking

df_categorized.corr()
```

```python
# unstack the matrix
correlation_mat = df_categorized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
# sort all pairs

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# define all r square values greater than 0.5 is high correlation
high_corr = sorted_pairs [(sorted_pairs) > 0.5]
high_corr
```
